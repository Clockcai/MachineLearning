## 算法实现

```python
# 引入scipy中的距离函数，默认为欧式距离
from scipy.spatial.distance import cdist

class K_Means(object):
    # 初始化参数，类别的个数k n_clusters(k)、迭代次数、初始化质心
    def __init__(self, n_clusters=7, max_iters=300, centroids=[]):
        self.n_clusters = n_clusters
        self.max_iters = max_iters
        # 将数组转化为np中的array
        self.centroids = np.array(centroids, dtype=np.float)
    # 训练样本
    def fit(self, data):
        # 判断质心的数组是否为空，为空的话给其赋初始值
        if(self.centroids.shape == (0, )):
            # 从data中随机找出k个点作为初始质心
            self.centroids = data[np.random.randint(0, data.shape[0], self.n_clusters), :]
            
        # 开始迭代
        for i in range(self.max_iters):
            # 计算距离矩阵， 得到一个（100， 6）的矩阵
            distances = cdist(data,self.centroids)
            # 在（500， 6）的矩阵的每一行中找出最小值，将最小值所在的列归为一类
            c_index = np.argmin(distances, axis=1)
            # 计算新的质心
            for i in range(self.n_clusters):
                # 如果质心点是原始数据以外的点，则可能出现有些质心并未有和数据中的其他点划分到一类
                # 因此需要做一个判断， 但是本例中的质心取自样本点，因此不存在此问题，他一定可以和自身聚为一类
                if i in c_index:
                    self.centroids[i] = np.mean(data[c_index == i],axis = 0)
    # 实现预测方法
    def predict(self, samples):
        # 计算点到质心的距离，选取距离质心距离最近的那个质心的类别
        distances = cdist(samples, self.centroids)
        # 计算每行最小值的下标， 按列返回
        c_index =np.argmin(distances, axis=1)
        return c_index
```

```python
# 定义一个绘图函数
def plotKMeans(x, y, centroids, subplot, title, c='r'):
    # 分配子图
    plt.subplot(subplot)
    plt.scatter(x[:,0], x[:,1], c=c)
    plt.scatter(centroids[:, 0], centroids[:, 1],c=np.array(range(7)), s=100)
    plt.title(title)
# 创建kmeans对象
kmeans = K_Means(max_iters=300, centroids=np.array([[2,1],[2,2],[2,3],[2,4],[2,5], [2, 6], [2, 7]]))
# 创建画布
plt.figure(figsize=(20, 8))
# 绘制初始的样本点
plotKMeans(x, y, kmeans.centroids, 121, 'Initial State')

# 开始聚类
kmeans.fit(x)
# 计算各点类别的矩阵
c_index = np.argmin(cdist(x,kmeans.centroids), axis=1)

# 将各个点用对应的类别的颜色进行标注
plotKMeans( x, y, kmeans.centroids, 122, 'Final State', c=c_index)

# 预测新数据对应的类别
x_new = np.array([[0, 0],[10, 7]])
y_pred = kmeans.predict(x_new)
# 聚类后的类别数组

print(kmeans.centroids)
print(y_pred)
print(c_index)

# c=y_pred表示用对应的类别进行标注颜色
plt.scatter(x_new[:,0], x_new[:,1], s=100, c=y_pred)
```
